```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
sns.get_dataset_names()
```

```python
data=sns.load_dataset("diamonds")
data.head(10)
```

```python
sns.barplot(x='carat', y='price', data=data.head(10))
```

```python
sns.relplot(x='carat', y='price', data=data.head(30))
```

```python
sns.lmplot(x='carat', y='price', data=data.head(30))
```

# summary:
the price increases with the carat
